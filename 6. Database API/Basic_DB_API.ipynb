# Create & Access SQLite database using Python


```python
# Create a database
import sqlite3
conn = sqlite3.connect('INSTRUCTORS.db')
cursor_obj = conn.cursor()

# Create a table
cursor_obj.execute("DROP TABLE IF EXISTS INSTRUCTORS")
table = """ create table IF NOT EXISTS INSTRUCTORS(ID INTEGER PRIMARY KEY NOT NULL, FNAME VARCHAR(20), LNAME VARCHAR(20), CITY VARCHAR(20), CCODE CHAR(2));"""
cursor_obj.execute(table)

# Insert into table
cursor_obj.execute('''insert into INSTRUCTORS values (1, 'Rav', 'Ahuja', 'TORONTO', 'CA')''')
cursor_obj.execute('''insert into INSTRUCTORS values (2, 'Raul', 'Chong', 'Markham', 'CA'), (3, 'Hima', 'Vasudevan', 'Chicago', 'US')''')
```

```python
# Query
statement = '''SELECT * FROM INSTRUCTORS'''
cursor_obj.execute(statement)
print("All the data")
output_all = cursor_obj.fetchall()

for row_all in output_all:
  print(row_all)
```

```python
import pandas as pd
#retrieve the query results into a pandas dataframe
df = pd.read_sql_query("select * from INSTRUCTORS;", conn)

#print the dataframe
df
```

```python
# Close the connection
conn.close()
```

# Analyzing a Real-World Data Set

```python
import prettytable # Display data in a formatted table
import csv, sqlite3
import pandas # Tool for data manipulation and analysis in Python

%load_ext sql # load the sql extension
con = sqlite3.connect("socioeconomic.db")
cur = con.cursor()
%sql sqlite:///socioeconomic.db # Establish a connection, connecting to magic sql using sqlite

df = pandas.read_csv('https://data.cityofchicago.org/resource/jcxq-k9xf.csv')
df.to_sql("chicago_socioeconomic_data", con, if_exists='replace', index=False,method="multi")
```

```python
prettytable.DEFAULT = 'DEFAULT' # sets the default style for tables
%sql SELECT * FROM chicago_socioeconomic_data limit 5;
```

```python
%sql SELECT COUNT(*) FROM chicago_socioeconomic_data;
```

```python
%sql SELECT COUNT(*) FROM chicago_socioeconomic_data WHERE hardship_index > 50.0;
```

```python
%sql SELECT MAX(hardship_index) FROM chicago_socioeconomic_data;
```

```python
%%sql

SELECT community_area_name
FROM chicago_socioeconomic_data
WHERE hardship_index = ( SELECT MAX(hardship_index) FROM chicago_socioeconomic_data );
```

```python
%sql SELECT community_area_name FROM chicago_socioeconomic_data WHERE per_capita_income_ > 60000;
```

```python
import seaborn as sns
income_vs_hardship = %sql SELECT per_capita_income_, hardship_index FROM chicago_socioeconomic_data;
plot = sns.jointplot(x='per_capita_income_',y='hardship_index', data=income_vs_hardship.DataFrame())
```

```python
import csv, sqlite3
import prettytable
import pandas

conn = sqlite3.connect("RealWorld.db")
cursor = conn.cursor()

prettytable.DEFAULT = 'DEFAULT'

%load_ext sql
%sql sqlite:///RealWorld.db

df = pandas.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DB0201EN-SkillsNetwork/labs/FinalModule_Coursera_V5/data/ChicagoPublicSchools.csv")
df.to_sql("CHICAGO_PUBLIC_SCHOOLS_DATA", conn, if_exists='replace', method='multi')
```

```python
%%sql
SELECT count(*)
FROM CHICAGO_PUBLIC_SCHOOLS_DATA
WHERE "Elementary, Middle, or High School"='ES'
```

```python
%%sql
SELECT MAX(SAFETY_SCORE) AS MAX_SAFETY_SCORE
FROM CHICAGO_PUBLIC_SCHOOLS_DATA
```

```python
%%sql
SELECT NAME_OF_SCHOOL, SAFETY_SCORE
FROM CHICAGO_PUBLIC_SCHOOLS_DATA
WHERE SAFETY_SCORE = (
  SELECT MAX(SAFETY_SCORE) AS MAX_SAFETY_SCORE
  FROM CHICAGO_PUBLIC_SCHOOLS_DATA
)
```

```python
%%sql
SELECT NAME_OF_SCHOOL, AVERAGE_STUDENT_ATTENDANCE
FROM CHICAGO_PUBLIC_SCHOOLS_DATA
ORDER BY AVERAGE_STUDENT_ATTENDANCE DESC
LIMIT 10
```

```python
%%sql
SELECT Name_of_School, Average_Student_Attendance
FROM CHICAGO_PUBLIC_SCHOOLS_DATA
ORDER BY Average_Student_Attendance
LIMIT 5
```

```python
%%sql
SELECT Name_of_School, REPLACE(Average_Student_Attendance, '%', '')
FROM CHICAGO_PUBLIC_SCHOOLS_DATA
ORDER BY Average_Student_Attendance
LIMIT 5
```

```python
%%sql
SELECT NAME_OF_SCHOOL, AVERAGE_STUDENT_ATTENDANCE
FROM CHICAGO_PUBLIC_SCHOOLS_DATA
WHERE CAST(REPLACE(Average_Student_Attendance, '%', '') AS DOUBLE ) < 70
ORDER BY AVERAGE_STUDENT_ATTENDANCE
```

```python
%%sql
SELECT community_area_name, sum(college_enrollment) as total_college_enrollment
FROM CHICAGO_PUBLIC_SCHOOLS_DATA
GROUP BY college_enrollment
```
