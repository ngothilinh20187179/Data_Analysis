```python
# Introduction to Matplotlib and Line Plots
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/Canada.csv')
df.head(2)
```

```python
df.set_index('Country', inplace=True)
df.index.name = None
df.head(2)
```

```python
print('Matplotlib version: ', mpl.__version__)
```

```python
print(plt.style.available)
mpl.style.use(['ggplot']) # optional: for ggplot-like style
```

```python
# Plot a line graph of immigration from Haiti
years = list(map(str, range(1980, 2014)))
haiti = df.loc['Haiti', years] # passing in years 1980 - 2013 to exclude the 'total' column
haiti.head()
```

```python
haiti.plot()
```

```python
haiti.plot(kind='line')
plt.title('Immigration from Haiti')
plt.ylabel('Number of immigrants')
plt.xlabel('Years')
plt.show()
```

```python
haiti.index
```

```python
haiti.values
```

```python
haiti.index = haiti.index.map(int)
haiti.plot(kind='line')

plt.title('Immigration from Haiti')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

# annotate the 2010 Earthquake.
# syntax: plt.text(x, y, label)
plt.text(2000, 6000, '2010 Earthquake') # toa do x, y

plt.show()
```

```python
# Plot a line graph of immigration from China and India
df_Chi_In = df.loc[['China', 'India'], years]
df_Chi_In
```

```python
df_Chi_In = df_Chi_In.transpose()
df_Chi_In.head()
```

```python
df_Chi_In.plot(kind='line')
plt.title('Immigrants from China and India')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

plt.show()
```

```python
# Top 5 countries - immigration to Canada.
df.head(5)
df.sort_values(by='Total', ascending=False, axis=0, inplace=True)
top_5 = df.head(5)
top_5
```

```python
top_5 = top_5[years]
top_5
```

```python
top_5 = top_5.transpose()
top_5
```

```python
top_5.plot(kind='line', figsize=(10, 8))
plt.title('Immigration Trend of Top 5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')
plt.show()
```
