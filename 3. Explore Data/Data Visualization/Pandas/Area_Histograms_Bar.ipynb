# Import, Fetching

```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

file_path = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/Canada.csv'
df = pd.read_csv(file_path)
df.head(2)
```

```python
df.shape
```

```python
df.set_index('Country', inplace=True)
df.head(2)
```

```python
df.shape
```

```python
years = list(map(str, range(1980, 2014)))
years
```

# Area Plots

```python
# Top 5 countries - immigrants to Canada
df.sort_values(by='Total', axis=0, ascending=False, inplace=True)
df_top5 = df.head(5)
df_top5
```

```python
df_top5 = df_top5[years]
df_top5 = df_top5.transpose()
df_top5
```

```python
# Option 1: Scripting layer (procedural method) - using matplotlib.pyplot as 'plt'
df_top5.plot(kind='area', stacked=False, figsize=(10, 5), alpha=0.35)
plt.title('Immigration Trend of Top 5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')
plt.show()
```

```python
# Option 2: Artist layer (Object oriented method) - using an Axes instance from Matplotlib (preferred)
ax = df_top5.plot(kind='area', alpha=0.55, stacked=False, figsize=(10, 5))
ax.set_title('Immigration Trend of Top 5 Countries')
ax.set_ylabel('Number of Immigrants')
ax.set_xlabel('Years')
```

# Histograms

```python
df['2013'].head()
```

```python
count, bin_edges = np.histogram(df['2013'])
print(count) # frequency count
print(bin_edges) # bin ranges, default = 10 bins
# 178 countries contributed between 0 to 3412.9 immigrants
# 11 countries contributed between 3412.9 to 6825.8 immigrants
# 1 country contributed between 6285.8 to 10238.7 immigrants, and so on..
```

```python
df['2013'].plot(kind='hist', figsize=(8, 4), xticks=bin_edges)
plt.title('Histogram of Immigration from 195 Countries in 2013')
plt.ylabel('Number of Countries')
plt.xlabel('Number of Immigrants')
plt.show()
```

```python
df.loc[['Denmark', 'Norway', 'Sweden'], years]
```

```python
df_t = df.loc[['Denmark', 'Norway', 'Sweden'], years].transpose()
df_t.head()
```

```python
count, bin_edges = np.histogram(df_t, 15)

df_t.plot(kind ='hist',
          figsize=(10,4),
          bins=15,
          alpha=0.6,
          xticks=bin_edges,
          color=['coral', 'darkslateblue', 'mediumseagreen']
         )

plt.title('Histogram of Immigration from Denmark, Norway, and Sweden from 1980 - 2013')
plt.ylabel('Number of Years')
plt.xlabel('Number of Immigrants')

plt.show()
```

```python
count, bin_edges = np.histogram(df_t, 15)
xmin = bin_edges[0] - 10   #  first bin value is 31.0, adding buffer of 10 for aesthetic purposes
xmax = bin_edges[-1] + 10  #  last bin value is 308.0, adding buffer of 10 for aesthetic purposes

# stacked Histogram
df_t.plot(kind='hist',
          figsize=(10, 4),
          bins=15,
          xticks=bin_edges,
          color=['coral', 'darkslateblue', 'mediumseagreen'],
          stacked=True,
          xlim=(xmin, xmax)
         )

plt.title('Histogram of Immigration from Denmark, Norway, and Sweden from 1980 - 2013')
plt.ylabel('Number of Years')
plt.xlabel('Number of Immigrants')

plt.show()
```

# Bar Charts

```python
df_iceland = df.loc['Iceland', years]
df_iceland.head()
```

```python
df_iceland.plot(kind='bar', figsize=(12, 4), rot=45)
plt.xlabel('Year')
plt.ylabel('Number of immigrants')
plt.title('Icelandic immigrants to Canada from 1980 to 2013')
# Annotate arrow
plt.annotate('',
             xy=(32, 70),
             xytext=(28, 20),
             xycoords='data',
             arrowprops=dict(arrowstyle='->', connectionstyle='arc3', color='blue', lw=2)
            )
# Annotate text
plt.annotate('2008 - 2011 Financial Crisis',
             xy=(28, 30),
             rotation=62,
             va='bottom',
             ha='left',
            )
plt.show()
```

```python
df.sort_values(by='Total', ascending=True, inplace=True)
df_top15 = df['Total'].tail(15)
df_top15
```

```python
df_top15.plot(kind='barh', figsize=(10, 8), color='steelblue')
plt.xlabel('Number of Immigrants')
plt.title('Top 15 Conuntries Contributing to the Immigration to Canada between 1980 - 2013')

for index, value in enumerate(df_top15):
    label = format(int(value), ',') # format int with commas
    plt.annotate(label, xy=(value - 47000, index - 0.10), color='white')

plt.show()
```
